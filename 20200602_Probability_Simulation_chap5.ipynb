```
#Jun-10-2020
#Python libraries need to generate random numbers and to graph chart
import numpy as np
import matplotlib.pyplot as plt
```

These two lines is to import the library numpy and matplotlib.

```
#generate ranodm numbers
#N is possible outcomes
#size is the size of the sample
#number_of_sample is the number of samples to be generated
def PsudoRand(N,size,number_of_sample):
    psudo = list(range(int(N)))#generate N unidform random number from uniform distribution
    choice = np.random.choice(psudo, (int(size),int(number_of_sample)))#choose N number from Rsudo without replacement
    return choice
```

```
#Crate Probabality charts
#N is possible outcomes
#size is the size of the sample
#number_of_sample is the number of samples to be generated
def Plot(N,size,number_of_sample):
  Sample = PsudoRand(N,size,number_of_sample)
  num_of_success = np.count_nonzero(Sample==0,axis=0)/int(size)
  plt.figure(figsize=(10,5))
  plt.plot(num_of_success)
  plt.axhline(y=1.0/N, color='r', linestyle='-')
  plt.title('Number of sample='+str(number_of_sample)+',Sample Size='+str(size)+',Possible outcomes='+str(N),fontsize=12)#set font here
  plt.suptitle('Probability chart',fontsize=16) 
  plt.ylim(0,1)
  plt.xlabel('Sample#')
  plt.ylabel('Success rate')
```

Two function defined above, PsudoRand is to simulate the game of tossing a dice or flip a coin. Plot is to plot the success of each sample
